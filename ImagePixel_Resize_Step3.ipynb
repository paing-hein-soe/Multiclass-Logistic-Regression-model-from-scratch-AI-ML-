```python
#%%
"""
Model: Multiclass Logistic Regression (Classficatio Model)  
@: paing_hein_soe
LinkedIn: https://www.linkedin.com/in/paing-hein-soe/

Step 3: Resize the images in the original folders to the target dimensions
"""
import os
import pandas as pd

# Define the target dimensions for resizing
pixel_width = 50
pixel_height = 50

# Load the folder statistics from the CSV file
image_categories = ["Wind Plant", "Solar Farm", "Thermal Plant"]
folder_path_parent="D:\ML_Archived\GoogleImage_LogReg"
csv_file_name = "image_data_statistics.csv" # To create this CSV file 

folder_name_items = image_categories
csv_file_path = os.path.join(folder_path_parent, csv_file_name)
folder_stats_loaded = pd.read_csv(csv_file_path, index_col=0)
print("Folder statistics loaded successfully.")
print(folder_stats_loaded)
```

```python
#%%
from PIL import Image
import os
# Define the original and output folder paths
original_folders = [folder_path_parent + "\\" + item for item in folder_name_items]
output_folders = [folder_path_parent + "\\" + item + "_resize" for item in folder_name_items]

#%%
# Iterate over each original folder
for original_folder, output_folder in zip(original_folders, output_folders):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Iterate over each image in the original folder
    for filename in os.listdir(original_folder):
        if filename.endswith('.jpg'):
            # Open the image
            img_path = os.path.join(original_folder, filename)
            img = Image.open(img_path)
            
            # Convert the image to RGB mode
            img = img.convert('RGB')
            
            # Resize the image
            resized_img = img.resize((pixel_width, pixel_height))
            
            # Save the resized image in the output folder
            output_path = os.path.join(output_folder, filename)
            resized_img.save(output_path)
            
            print(f"Image {filename} resized and saved to {output_path}")

print("All images resized and saved successfully.")



# %%
```
