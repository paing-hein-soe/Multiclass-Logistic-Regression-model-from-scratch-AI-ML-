```python
image_categories = ["Wind Plant", "Solar Farm", "Thermal Plant"]
folder_path_parent="D:\ML_Archived\GoogleImage_LogReg"
csv_file_name = "image_data_statistics.csv" # To create this CSV file 
images_download = 200 # Number of images to download (MAXIMUM - Google doesn't allow freely)
```

```python

# (Above) Replace with your search query and desired save path
for ii in range(len(image_categories)):
    search_query_1 = image_categories[ii]
    folder_path_parent_1 = folder_path_parent + "\\" + search_query_1
    save_path = folder_path_parent_1
    search_query = search_query_1
    # Configure and start the crawler
    google_crawler = MyGoogleImageCrawler(
        feeder_threads=1,  # Number of threads for feeder
        parser_threads=2,  # Number of threads for parser
        downloader_threads=4,  # Number of threads for downloader
        storage={"root_dir": save_path}, # Save path for downloaded images
    )
    
    google_crawler.crawl(keyword=search_query, max_num=images_download)  # Number of images to download
    
                
    for filename in os.listdir(save_path):
        if filename.lower().endswith('.png'):
            png_path = os.path.join(save_path, filename)
            png_image = Image.open(png_path)
            
            # Remove ".png" extension and add ".jpg" extension for output filename
            jpg_filename = os.path.splitext(filename)[0] + ".jpg"
            jpg_path = os.path.join(save_path, jpg_filename)
            
            # Convert and save as JPG
            png_image.convert("RGB").save(jpg_path, "JPEG")
            print(f"Converted: {filename} -> {jpg_filename}")
    
    valid_extensions = ['.jpg']
    for root, _, files in os.walk(save_path):
        for filename in files:
            if os.path.splitext(filename)[1].lower() not in valid_extensions:
                os.remove(os.path.join(root, filename))
                print(f"Removed: {filename} (Invalid extension)")
    
```
